**Univariate Analysis**

---



Non-Visual Analysis

```python
def numerical_data(numerical_feature):
  for col in numerical_feature:
    print('*'*5, col, '*'*5)
    print(numerical_feature[col].agg(['count', 'nunique', 'mean','max','min','std','skew','kurtosis']))
    print()
```

```python
df = data[['calories', 'intensity', 'steps', 'mets']]
```

```python
numerical_data(df)
```

```python
fig ,axes = plt.subplots(nrows=int(np.ceil(len(df.columns)/2)),
ncols=2,figsize=(10,12))
for i, column in enumerate(df.columns):
  row = int(i//2)
  col = i%2
  sns.distplot(df[column],ax=axes[row,col])
  fig.suptitle("Distribution Plots")

plt.show()
```

```python
df.plot(kind='box',subplots=True,layout=(8,3),figsize=(12,16))
```

Univariate Analysis Observation:

- All the features in data are right skewed.
- All the features in data have outliers.

**Bivariate Analysis**

---



```python
f,ax = plt.subplots(figsize=(10,8))
corr = data.corr()
hm = sns.heatmap(round(corr,2),annot=True, cmap='viridis',fmt='.2f')
t=f.suptitle('Heatmap',fontsize=20)
```

- Calories. intensity, steps and mets are highly correlated.

```python
fig,axes = plt.subplots(1,4,figsize=(12,4))
data.plot(kind='scatter', x='intensity', y='calories',ax=axes[0])
axes[0].set_title('Calories by Intensity')

data.plot(kind='scatter', x='steps',y='calories', ax=axes[1])
axes[1].set_title('Calories by Steps ')

data.plot(kind='scatter', x='intensity', y='steps',ax=axes[2])
axes[2].set_title('Steps by Intensity')

data.plot(kind='scatter', x='steps', y='mets',ax=axes[3])
axes[3].set_title('Steps by METs')

plt.tight_layout()
plt.show()
```

- With intensity 1 or 3 steps per minute are above 150.
- Thus, intensity 1 or 3 can burn calories above 17.

```python
data_steps_id = data.groupby(['id'])['steps'].mean()
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(x=data_steps_id.index, y=data_steps_id.values,color='salmon')
plt.xlabel('User ID')
plt.ylabel('Average of Total Steps')
plt.title('Average Steps of  User')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Minimum steps per minute by most of the users is around 4.

```python
data_intensity_id = data.groupby(['id'])['intensity'].mean()
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(x=data_intensity_id.index, y=data_intensity_id.values,color='skyblue')
plt.xlabel('User ID')
plt.ylabel('Average of Total Intensity')
plt.title('Average Intensity of  User')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- The minimum average intensity per minute of most of the users is 0.10.

```python
data_calories_id = data.groupby(['id'])['calories'].mean()
```

```python
plt.figure(figsize=(8, 8))
sns.barplot(x=data_calories_id.index, y=data_calories_id.values,color='salmon')
plt.xlabel('User ID')
plt.ylabel('Average Calories')
plt.title('Average Calories of  User')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Minimum average calories per minute of all users is 1.

Overall Observations:
- With intensity 1 or 3 steps per minute are above 150.
- The intensity 1 or 3 can burn calories above 17.
- Minimum steps per minute by most of the users is around 4.
- The minimum average intensity per minute of most of the users is 0.10.
- Minimum average calories per minute of all users is 1.
